```python
from sklearn.externals import joblib
clf = joblib.load('../../../data/models/svm.model')
import cv2
from matplotlib import pyplot as plt

import os 
imgs = os.listdir()
imgs = [i for i in imgs if '.jpg' in i]
hog = cv2.HOGDescriptor((48,48),(16,16),(8,8),(8,8),9)
import numpy as np
from PIL import Image
for i in imgs:
    im = Image.open(i)
    im = im.resize((120,120))

    im = np.array(im)
    im = cv2.cvtColor(im,cv2.COLOR_RGB2BGR)
    fd = hog.compute(im , (120,120))
    fd = fd.reshape(1,-1)
    pre = clf.predict(fd)
    plt.imshow(im)
    plt.title(pre)
    plt.pause(0.01)
    print(clf.decision_function(fd))
```
